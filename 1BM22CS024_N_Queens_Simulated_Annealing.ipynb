```python
initial = [3,1,2,0]
goal_val = 0
```

```python
def calculateCost(state):
  cost = 0
  size = len(state)
  for i in range(size):
    for j in range(i+1,size):
      if abs(state[i]-state[j]) == abs(i-j) or state[i] == state[j]:
        cost+=1
  return cost
```

```python
def getNeighbours(state):
  size = len(state)
  curr_state = state[:]
  neighbours = []
  for i in range(size):
    for j in range(i+1,size):
       curr_state[j], curr_state[i] = state[i], state[j]
       neighbours.append(curr_state)
       curr_state = state[:]
  return neighbours
```

```python
def printBoard(state):
  size = len(state)
  board = [['_' for i in range(size)] for i in range(size)]
  for i in range(0,size):
    board[state[i]][i] = 'Q'
  display(board)
```

```python
import numpy as np
from math import exp

def simAnnealing(initial, goal_val, t):
  curr_state = initial[:]
  T = t

  while(T > 0):
    neighbours = getNeighbours(curr_state)
    next_index = np.random.choice([i for i in range(0,len(neighbours))])
    next = neighbours[next_index]
    next_cost = calculateCost(next)
    curr_cost = calculateCost(curr_state)
    if curr_cost == goal_val:
      print("Solution reached")
      printBoard(curr_state)
      return
    del_E = curr_cost - next_cost
    p = 1
    selected = False
    if del_E > 0:
      curr_state = next
      selected = True
    else:
      p = exp(del_E)/T
      index = np.random.choice([0,1],p=[p, 1-p])
      if index==0:
        curr_state = next
        selected = True
    print("Cost: ",curr_cost," next cost: ",next_cost," T: ", T, "p: ",p,"next selected: ",selected)
    T -= 0.5
  print("Could not reach global optima: ")
  printBoard(curr_state)
```

```python
simAnnealing([6,4,0,1,2,3,5,7],goal_val,100)
```
